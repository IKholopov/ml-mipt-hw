```python
from sklearn.tree import DecisionTreeClassifier
```

### Загрузка датасета

```python
import pandas as pd
data = pd.read_csv('german.data-numeric', header=None, delim_whitespace=True)

X = data[list(set(data.columns) - set([24]))]
y = data[24]
```

### Построение дерева

#### Без ограничения глубины

```python
import pydotplus
from IPython.display import Image
import sklearn.tree

class_names = ['Good', 'Bad']

def BuildAndVisualizeTree(X, y, max_depth=None):
    tree = DecisionTreeClassifier(max_depth=max_depth)
    tree.fit(X, y)
    tree_data = sklearn.tree.export_graphviz(tree, out_file='tree.out',
                                             class_names=class_names, filled=True,
                                             rounded=True, special_characters=True)
    graph = pydotplus.graphviz.graph_from_dot_file('tree.out')
    image = Image(graph.create_png())
    return tree, image
```

```python
tree, image = BuildAndVisualizeTree(X, y)
```

```python
image
```

Как видно, дерево получилось глубоким и скорее всего переоученным. Попробуем ограничить глубину и проинтерпретировать результат.

```python
tree, image = BuildAndVisualizeTree(X, y, 2)
image
```

Если на счет клиента приходила зарплата в более 200 марок в год или счёт отсутствует, то он получит кредит.
Иначе, смотрим на сколько месяцев берётся кредит. Если не более 22 - даём кредит, иначе нет.

Посчитаем качество на кросс-валидации и на обучающей выборке

```python
from sklearn.cross_validation import cross_val_score
import matplotlib.pyplot as plt
%matplotlib inline

acc_score = []
for depth in range(2, 20):
    tree = DecisionTreeClassifier(max_depth=depth)
    tree.fit(X, y)
    acc_score.append(cross_val_score(tree, X, y, scoring='accuracy').mean())
```

```python
def DepthPlot(acc_score, title):
    plt.title(title)
    plt.xlabel('Depth')
    plt.ylabel('Accuracy')
    plt.plot(range(2, 20), acc_score)
    plt.show()
```

```python
DepthPlot(acc_score, 'CV')
```

```python
from sklearn.metrics import accuracy_score

acc_score = []
for depth in range(2, 20):
    tree = DecisionTreeClassifier(max_depth=depth)
    tree.fit(X, y)
    acc_score.append(accuracy_score(y, tree.predict(X)))
```

```python
DepthPlot(acc_score, 'Training dataset')
```

Как видно при сравнении графиков, после глубины 5 деревья начинают сильно переобучаться. Поэтому глубокие деревья не очень выгодно использовать без дополнительной обработки.
